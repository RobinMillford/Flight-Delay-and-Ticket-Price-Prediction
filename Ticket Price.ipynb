## Duration

```python
def convert_duration(Duration):
        if len(Duration.split()) == 2:
            hours = int(Duration.split()[0][: -1])
            minutes = int(Duration.split()[1][: -1])
            return hours * 60 + minutes
        else:
            return int(Duration[: -1]) * 60
```

```python
df['Duration'] = df['Duration'].apply(convert_duration)
df.head()
```

# Building the Machine Learning Models and Evaluatiing them

```python
temp_col = df.columns.to_list()
print(temp_col,'\n')

new_col = temp_col[:2]+temp_col[3:]
new_col.append(temp_col[2])
print(new_col,'\n')

df = df.reindex(columns=new_col)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler =  StandardScaler()

df = scaler.fit_transform(df)

df[0]
```

```python
from sklearn.model_selection import train_test_split as tts

x = df[: , : -1]
y = df[: , -1]
```

```python
x_train,x_test,y_train,y_test = tts(x,y,test_size=0.1,random_state=42)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

def metrics(y_true, y_pred):
    print('RMSE:', mean_squared_error(y_true, y_pred)**0.5)
    print('R-squared:', r2_score(y_true, y_pred))

def accuracy(y_true,y_pred):
    errors = abs(y_pred-y_true)
    mape = 100*np.mean(errors/y_true)
    accuracy = 100 - mape
    return accuracy
```

```python
y_pred = model.predict(x_test)
```

```python
metrics(y_test,y_pred)
```

```python
accuracy(y_test,y_pred)
```

## Rnadom Forest

```python
from sklearn.ensemble import RandomForestRegressor

model_random_forest = RandomForestRegressor(n_estimators=100, min_samples_split=3)
model_random_forest.fit(x_train, y_train)
```

```python
y_pred_rf = model_random_forest.predict(x_test)
```

```python
metrics(y_test,y_pred_rf)
```

```python
accuracy(y_test,y_pred_rf)
```
